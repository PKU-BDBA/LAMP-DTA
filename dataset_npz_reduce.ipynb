```python
max_size = 1024
filled_pid2llm = dict()
for key in pid2llm:
    x = pid2llm[key][0]
    features = x.shape[1]
    out = np.zeros([max_size, features])
    out[:x.shape[0], :] = x
    filled_pid2llm[key] = out
```

```python
step = 8
reduced_pid2llm = dict()
for key in filled_pid2llm:
    x = filled_pid2llm[key]
    x1 = np.zeros([x.shape[0] // step, x.shape[1]])
    # x2 = np.zeros([x.shape[0] // step, x.shape[1] // step])

    for i1, i in enumerate(np.arange(x.shape[0], step=step)):
        x1[i1: i1 + 1, :] = np.mean(x[i: i + step, :], axis=0, keepdims=True)

    # for j1, j in enumerate(np.arange(x.shape[1], step=step)):
        # x2[:, j1: j1 + 1] = np.mean(x1[:, j: j + step], axis=1, keepdims=True)
    # reduced_pid2llm[key] = x2
    reduced_pid2llm[key] = x1
```

```python
np.savez(dataset + '_reduced.npz', dict=reduced_pid2llm)
```
